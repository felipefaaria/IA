```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
import pickle
try:
    base = pd.read_csv('train.csv')
    print("DataFrame do Titanic carregado com sucesso.")
except FileNotFoundError:
    print("Erro: O arquivo 'train.csv' não foi encontrado. Por favor, certifique-se de que ele foi enviado.")
    raise
    
base['Age'].fillna(base['Age'].median(), inplace=True)
base['Embarked'].fillna(base['Embarked'].mode()[0], inplace=True)

# Remover colunas que não serão usadas
base.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Codificar a coluna 'Sex' (binária) com LabelEncoder
base['Sex'] = LabelEncoder().fit_transform(base['Sex'])

# Codificar a coluna 'Embarked' (não ordinal) com OneHotEncoder
base = pd.get_dummies(base, columns=['Embarked'], prefix='Embarked', drop_first=True)

print("DataFrame após pré-processamento:")
print(base.head())
print("\nInformações finais:")
base.info()
```
